## หัวข้อโครงงานที่สนใจ 

### ชื่อหัวข้อ

ระบบจำแนกอารมณ์ของข้อความทวีตภาษาอังกฤษ (Basic Emotion Classification in English Tweets)



###  ประโยขน์ของโมเดล

* ช่วยในการ วิเคราะห์อารมณ์ของผู้ใช้สื่อสังคมออนไลน์ เช่น Twitter หรือ Facebook
* สนับสนุนการทำ Social Media Monitoring สำหรับองค์กร เพื่อตรวจจับความคิดเห็นเชิงบวก-ลบ-อารมณ์
* ใช้ต่อยอดใน ระบบตอบกลับอัตโนมัติ (Chatbot) ที่เข้าใจอารมณ์ผู้ใช้ได้
* ใช้ในการทำ วิเคราะห์แนวโน้ม (Trend Analysis) ของสังคมหรือเหตุการณ์สำคัญ
* สามารถนำไปพัฒนาเป็น เครื่องมือวิเคราะห์ความคิดเห็นลูกค้า (Customer Feedback Emotion Analysis)
* เป็นพื้นฐานในการสร้างระบบ AI ที่เข้าใจอารมณ์มนุษย์ (Emotion-aware AI)

###  ข้อมูลในการทำโครงงาน


*   Dataset: Emotions in Text Dataset
* Dataset: https://www.kaggle.com/datasets/ishantjuyal/emotions-in-text
*   Description:
ข้อมูลชุดนี้ประกอบด้วยข้อความทวีตจากผู้ใช้ Twitter พร้อมป้ายกำกับอารมณ์ (Emotion Labels)
โดยแต่ละตัวอย่างประกอบด้วย
   * tweet_text : ข้อความในทวีต (Text Data)
   * emotion : อารมณ์ของข้อความ เช่น
       * joy
       * sadness
       * anger
       * fear
       * love
       * surprise

* จำนวนข้อมูลประมาณ 20,000–40,000 ตัวอย่าง (แล้วแต่เวอร์ชันของชุดข้อมูล)
* Data Source: https://www.kaggle.com/code/damlongrat/emotion-classification
